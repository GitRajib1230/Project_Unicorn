```python
import pandas as pd


# Define a function to handle currency and amount extraction
def split_currency_amount(value):
    if pd.isna(value):  # Handle missing values
        return pd.Series([None, None])
    
    # Ensure the value is a string
    value = str(value)
    
    # List of possible currency symbols
    currency_symbols = ['$', '£', '€', '₹', '₩', 'CN¥']
    
    # Check which symbol is in the string
    for symbol in currency_symbols:
        if value.startswith(symbol):
            # Remove the currency symbol and commas, then convert to float
            amount = value.strip(symbol).replace(',', '')
            return pd.Series([symbol, float(amount)])
    
    # If no symbol matches, return None
    return pd.Series([None, None])

# Apply the function to split the column
clean[['Currency_E', 'Ttl Equity Funding Amount']] = clean['Total Equity Funding Amount'].apply(split_currency_amount)
clean[['Currency_T', 'Ttl Funding Amount']] = clean['Total Funding Amount'].apply(split_currency_amount)

clean
```

```python
# Define Currency Conversion Rates

conversion_rates = {
    '$': 1,         # USD to USD
    '¥': 0.1380,      # CNY to USD 
    '€': 1.1,        # EUR to USD
    '₹': 0.01184,    #Indian Rupee
    '₩': 0.0007126     #South Korean Won
}
```

```python
# Convert to USD

clean['Ttl Equity Funding Amount in USD'] = clean.apply(
    lambda row: row['Ttl Equity Funding Amount'] * conversion_rates.get(row['Currency_E'], 1),
    axis=1
)

clean['Ttl Funding Amount in USD'] = clean.apply(
    lambda row: row['Ttl Funding Amount'] * conversion_rates.get(row['Currency_T'], 1),
    axis=1
)
```
